# Colab Basics
This is a living document intended to get new users up and running with Google colab. These are just the basics of starting a runtime, checking the environment, and connecting with Google Drive

Setup:

1) Make sure GPU is enabled, go to colab's **edit** menu (not your browser's) edit->notebook settings->hardware accelerator: GPU

2) If this notebook is already in your Google drive, disregard. If you accessed it from a weblink, make a copy in your personal drive, file->Save a copy in Drive

3) Running your first code cell will start the runtime. Press the play arrow next to !nvidia-smi (this will fail if you haven't enabled GPU)

**Line 1)** `!nvidia-smi` will report your runtime's GPU environment including CUDA version and hardware (typically Tesla)

```
!nvidia-smi
```

**Line2)** Your runtime will come loaded with many typical data science libraries. Running `import tensorflow` will import the default TF version (currently 2.x). You can use 1.x by running a cell with  `tensorflow_version 1.x`  **before** you run import tensorflow.

```
%tensorflow_version 1.x
```

Line 3) Use `pip` to list all of the installed packages (and version #s) on your runtime.

```
pip list
```

Line 4) Look at the three icons on the far left bar -> content, code, and files. Press **files** to open up your runtime's file structure. You will start in the system's **content** folder. To confirm this, download a small file from the internet (a .txt of from Project Gutenberg). Use `wget` to download, and it will show in content (may need to click refresh in the files tab)

```
!wget https://www.gutenberg.org/files/98/98-0.txt
```

Line 5) The git code linked below is currently broken, but as an example, this line will clone any git repository into your content folder.

```
!git clone https://github.com/tivonrice/gpt-2.git
```

Line 6) Navgating directories is the same as always

```
cd gpt-2
```

Line 7) It looks like COLAB may have added some GUI elements in the Files tab to download, upload, or mount Google Drive. Otherwise, the following line will mount drive for saving and accessing checkpoints later. Follow the link to log in to your google account, then copy your auth code into the interactive field below.


```
from google.colab import drive
drive.mount('/content/drive')
```

Line 8) cd into your drive

```
cd /content/drive
```

Line 9) now you can use simple `!cp -r` commands to move things in and out of your Google drive. The following will copy the Gutenberg text from your runtime to your personal Google drive's root folder.

```
!cp -r /content/98-0.txt /content/drive/My\ Drive/
```

Line 10) when you start running .py scripts in COLAB (either from a git or your drive, make sure to add a `!` such as below (fake, do not execute)

```
!python3 my_brilliant_model.py
```
